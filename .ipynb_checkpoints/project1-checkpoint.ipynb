```python
def extract_ep(txt):
    ep_count=""
    checked_started = False
    txt_size= len(txt)
    for i in range (txt_size):
        if txt[i] == '(':
            checked_started =True
        if txt[i] == ')':
            checked_started=False
            return ep_count
        if checked_started == True:
            if txt[i] != ')' and txt[i] != '(' :
                ep_count = ep_count+ txt[i]
```

```python
extract_ep('Fullmetal Alchemist: BrotherhoodTV (64 eps)Apr 2009 - Jul 20103,218,472 membersManga StoreVolume 1â¬4.58Preview')
```

```python
df["Episode Count"]=df['Title'].apply(extract_ep)
```

```python
df["Episode Count"]=df["Episode Count"].str.replace(" eps","")
```
